```python
# Uploading the dataset from local system
from google.colab import files
upload = files.upload()

# Importing necessary libraries
import pandas as pd                  # For data handling
import seaborn as sns               # For visualization (optional)
import numpy as np                  # For numerical operations
import matplotlib.pyplot as plt     # For plotting
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
# Load CSV file into DataFrame
df = pd.read_csv('IRIS.csv')
df.head()  # Show first few rows
```

```python
# Removing duplicate rows from the dataset
df = df.drop_duplicates()

# Optional: Check how many duplicates were originally present
# print("Duplicate rows found:", df.duplicated().sum())
```

```python
# Generating descriptive statistics for numerical features
df.describe()
```

```python
# Selecting features (input variables)
x = df[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]

# Selecting target variable (class labels)
y = df['species']
```

```python
# Encode class labels into numeric values
from sklearn.preprocessing import LabelEncoder, StandardScaler
label_encoder = LabelEncoder()
y_integer = label_encoder.fit_transform(y)
```

```python
# Convert labels to one-hot encoded format (multi-class requirement)
from tensorflow.keras.utils import to_categorical
y_encoded = to_categorical(y_integer)
```

```python
# Standardize feature values
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

```python
# Splitting dataset into training and testing
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(
    x_scaled, y_encoded, test_size=0.2, random_state=42
)
```

```python
# Building the ANN model
model = Sequential()

# First hidden layer
model.add(Dense(32, input_shape=(4,), activation='relu'))

# Second hidden layer
model.add(Dense(8, activation='relu'))

# Output layer for 3-class classification
model.add(Dense(3, activation='softmax'))
```

```python
# Compile the model with appropriate loss function and optimizer
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
# Train the ANN model
model.fit(x_train, y_train, epochs=100, batch_size=8, validation_data=(x_test, y_test))
```
